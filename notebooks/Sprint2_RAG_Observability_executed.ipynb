**Purpose**: Observe and inspect the RAG system ingestion pipeline:

This notebook demonstrates all Sprint 2 deliverables through API observation.

## Setup and Imports

```python
import sys
sys.path.append('../src')

import requests
import json
from datetime import datetime
import pandas as pd

print(f"📊 Sprint 2 RAG Observability Started: {datetime.now()}")
print("🎯 Observing: Embedding execution, ingestion, vector DB operations")
API_BASE = "http://127.0.0.1:8000"
```

## 1. Test API Connectivity

```python
print("=== API CONNECTIVITY TEST ===")
try:
    response = requests.get(f"{API_BASE}/health")
    print(f"✅ API Status: {response.status_code}")
    health_data = response.json()
    print(f"✅ Server Version: {health_data.get('python_version')}")
    print(f"✅ Dependencies: {health_data.get('dependencies_loaded')}")
except Exception as e:
    print(f"❌ API not running. Start server with:")
    print(f"   uv run uvicorn src.main:app --reload")
    print(f"   Error: {e}")
```

## 2. Test Embedding Model Execution

```python
print("=== EMBEDDING MODEL EXECUTION TEST ===")
response = requests.get(f"{API_BASE}/vector-db/test-embedding")
print(f"Status Code: {response.status_code}")

if response.status_code == 200:
    embedding_test = response.json()
    print("✅ Embedding Model Test Results:")
    print(f"   Model: {embedding_test.get('model')}")
    print(f"   Status: {embedding_test.get('status')}")
    print(f"   Message: {embedding_test.get('message')}")
    print(f"   Test Text: {embedding_test.get('test_text')[:50]}...")
else:
    print(f"❌ Embedding test failed: {response.text}")
```

## 3. Initial Vector Database Status

```python
print("=== INITIAL VECTOR DB STATUS ===")
response = requests.get(f"{API_BASE}/vector-db/status")
initial_status = response.json()

print(f"📊 Database Status: {initial_status.get('status')}")
print(f"📁 Documents: {initial_status.get('num_documents')}")
print(f"🔗 Nodes: {initial_status.get('num_nodes')}")
print(f"✅ Index Ready: {initial_status.get('index_ready')}")
print(f"🔧 Embedding Model: {initial_status.get('embedding_model')}")
print(f"🧠 LLM Model: {initial_status.get('llm_model')}")
print(f"📦 Chunk Size: {initial_status.get('chunk_size')}")
print(f"💾 Storage: {initial_status.get('storage_directory')}")
```

## 4. Wikipedia Data Ingestion & Embedding Generation

```python
print("=== WIKIPEDIA INGESTION & EMBEDDING GENERATION ===")
print("🚀 Starting Wikipedia data ingestion...")
start_time = datetime.now()

# Configure ingestion parameters
ingestion_payload = {
    "force_rebuild": True,
    "chunk_size": 512
}

print(f"⚙️ Configuration: {ingestion_payload}")
print("📡 Sending ingestion request...")

# Trigger ingestion
response = requests.post(f"{API_BASE}/ingest", json=ingestion_payload)
end_time = datetime.now()
duration = end_time - start_time

print(f"⏱️ Total ingestion time: {duration}")
print(f"📊 HTTP Status Code: {response.status_code}")

if response.status_code == 200:
    result = response.json()
    print("\n✅ INGESTION SUCCESSFUL!")
    print("📋 Results:")
    print(f"   Status: {result.get('status')}")
    print(f"   Documents Loaded: {result.get('documents_loaded')}")
    print(f"   Embedding Model: {result.get('embedding_model')}")
    print(f"   Vector DB Ready: {result.get('vector_db_ready')}")
    print(f"   Message: {result.get('message')}")
else:
    print(f"❌ INGESTION FAILED!")
    print(f"   Error: {response.text}")
```

## 5. Final Vector Database Status & Comparison

```python
print("=== FINAL VECTOR DB STATUS & COMPARISON ===")
response = requests.get(f"{API_BASE}/vector-db/status")
final_status = response.json()

print("📈 FINAL Status:")
print(f"   Status: {final_status.get('status')}")
print(f"   Documents: {final_status.get('num_documents')}")
print(f"   Nodes: {final_status.get('num_nodes')}")
print(f"   Index Ready: {final_status.get('index_ready')}")
print(f"   Storage: {final_status.get('storage_directory')}")

# Calculate and display changes
print(f"\n🔄 CHANGES FROM INGESTION:")
docs_before = initial_status.get('num_documents', 0)
docs_after = final_status.get('num_documents', 0)
docs_added = docs_after - docs_before

print(f"   Documents: {docs_before} → {docs_after} (+{docs_added})")
print(f"   Index Ready: {initial_status.get('index_ready')} → {final_status.get('index_ready')}")

if docs_added > 0:
    print(f"✅ Successfully added {docs_added} Wikipedia documents to vector database!")
else:
    print("⚠️ No new documents were added.")
```
