## 1. Environment Setup Verification

```python
import sys
import os
import requests
import json
from datetime import datetime

print(f"Python version: {sys.version}")
print(f"Current working directory: {os.getcwd()}")
print(f"Validation timestamp: {datetime.now().isoformat()}")
```

## 2. FastAPI Dependencies Check

```python
try:
    import fastapi
    import uvicorn
    import pydantic
    
    print("✅ FastAPI Dependencies Successfully Imported:")
    print(f"  - FastAPI: {fastapi.__version__}")
    print(f"  - Uvicorn: {uvicorn.__version__}")
    print(f"  - Pydantic: {pydantic.__version__}")
except ImportError as e:
    print(f"❌ Import Error: {e}")
```

## 3. Server Health Check

```python
# Test server is running and accessible
SERVER_URL = "http://localhost:8000"

try:
    response = requests.get(f"{SERVER_URL}/health", timeout=5)
    if response.status_code == 200:
        health_data = response.json()
        print("✅ Server Health Check PASSED")
        print(json.dumps(health_data, indent=2))
    else:
        print(f"❌ Health check failed with status: {response.status_code}")
except requests.exceptions.RequestException as e:
    print(f"❌ Server connection failed: {e}")
    print("   Make sure FastAPI server is running on port 8000")
```

## 4. Root Endpoint Validation

```python
try:
    response = requests.get(f"{SERVER_URL}/", timeout=5)
    if response.status_code == 200:
        root_data = response.json()
        print("✅ Root Endpoint PASSED")
        print(f"  - API Name: {root_data.get('message')}")
        print(f"  - Version: {root_data.get('version')}")
        print(f"  - Documentation: {root_data.get('docs')}")
        print(f"  - Health Check: {root_data.get('health')}")
    else:
        print(f"❌ Root endpoint failed with status: {response.status_code}")
except requests.exceptions.RequestException as e:
    print(f"❌ Root endpoint connection failed: {e}")
```

## 5. Echo Endpoint Validation

```python
# Test echo endpoint functionality
test_message = "Sprint 1 validation from Jupyter notebook"
test_metadata = {"notebook": "01_data_exploration", "sprint": 1}

try:
    response = requests.post(
        f"{SERVER_URL}/echo", 
        json={"message": test_message, "metadata": test_metadata},
        timeout=5
    )
    if response.status_code == 200:
        echo_data = response.json()
        print("✅ Echo Endpoint PASSED")
        print(f"  - Echoed Message: {echo_data.get('echo')}")
        print(f"  - Metadata Received: {echo_data.get('received_metadata')}")
        print(f"  - Timestamp: {echo_data.get('timestamp')}")
    else:
        print(f"❌ Echo endpoint failed with status: {response.status_code}")
except requests.exceptions.RequestException as e:
    print(f"❌ Echo endpoint connection failed: {e}")
```

## 6. API Documentation Accessibility

```python
# Test documentation endpoints
docs_endpoints = ["/docs", "/redoc"]

for endpoint in docs_endpoints:
    try:
        response = requests.get(f"{SERVER_URL}{endpoint}", timeout=5)
        if response.status_code == 200:
            print(f"✅ {endpoint} endpoint accessible")
            print(f"  - Content-Type: {response.headers.get('content-type')}")
            print(f"  - Content-Length: {response.headers.get('content-length')} bytes")
        else:
            print(f"❌ {endpoint} failed with status: {response.status_code}")
    except requests.exceptions.RequestException as e:
        print(f"❌ {endpoint} connection failed: {e}")
```
