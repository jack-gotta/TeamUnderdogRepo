## 4. Endpoint Validation

```python
# Test Health Check Endpoint
print("=== HEALTH CHECK ENDPOINT ===")
response = client.get("/health")
print(f"Status Code: {response.status_code}")
print(f"Response: {response.json()}")

if response.status_code == 200:
    data = response.json()
    if data.get('status') == 'healthy':
        print("✅ Health check PASSED")
    else:
        print("❌ Health check reports unhealthy")
else:
    print("❌ Health check FAILED")
```

```python
# Test Echo Endpoint (POST)
print("=== ECHO ENDPOINT (POST) ===")
test_message = "Sprint 1 Validation Test"
response = client.post("/echo", json={"message": test_message})
print(f"Status Code: {response.status_code}")
print(f"Response: {response.json()}")

if response.status_code == 200:
    data = response.json()
    if data.get('echo') == test_message:
        print("✅ Echo POST endpoint PASSED")
    else:
        print("❌ Echo response doesn't match input")
else:
    print("❌ Echo POST endpoint FAILED")
```

```python
# Test Echo Endpoint (GET)
print("=== ECHO ENDPOINT (GET) ===")
test_message = "get-test-123"
response = client.get(f"/echo/{test_message}")
print(f"Status Code: {response.status_code}")
print(f"Response: {response.json()}")

if response.status_code == 200:
    data = response.json()
    if data.get('echo') == test_message:
        print("✅ Echo GET endpoint PASSED")
    else:
        print("❌ Echo GET response doesn't match input")
else:
    print("❌ Echo GET endpoint FAILED")
```

```python
# Test FastAPI Docs Endpoint
print("=== FASTAPI DOCS ENDPOINT ===")
response = client.get("/docs")
print(f"Status Code: {response.status_code}")
print(f"Content-Type: {response.headers.get('content-type')}")

if response.status_code == 200:
    content = response.text
    if "swagger" in content.lower() and "openapi" in content.lower():
        print("✅ FastAPI docs endpoint PASSED")
        print("   Contains Swagger UI elements")
    else:
        print("❌ Docs don't contain expected Swagger content")
else:
    print("❌ FastAPI docs endpoint FAILED")
```

```python
# Test Root Endpoint
print("=== ROOT ENDPOINT ===")
response = client.get("/")
print(f"Status Code: {response.status_code}")
print(f"Response: {response.json()}")

if response.status_code == 200:
    data = response.json()
    if "message" in data and "version" in data:
        print("✅ Root endpoint PASSED")
    else:
        print("❌ Root endpoint missing required fields")
else:
    print("❌ Root endpoint FAILED")
```
